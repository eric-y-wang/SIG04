```python
import numpy as np
import scanpy as sc
import anndata as ad
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

# Import Data

```python
adata = ad.read_h5ad("/data/rudensky/EYW/SIG04_240816/python_outs/SIG04_merge_singlets_ZscoreNorm_scanpy_clustered.h5ad")

# select lane 2 only
adata = adata[adata.obs['gem_group']=="lane2"].copy()
adata.X = adata.layers['log1p_norm'].copy()
adata.layers["scaled"] = sc.pp.scale(adata, copy=True).X

# remove pooled samples
adata = adata[~adata.obs['oBC_feature_call'].str.contains('-pool')].copy()
```

```python
deg = pd.read_csv("/data/rudensky/EYW/git_projects/SIG04_240816/analysis_outs/DEG_MAST_lane2.csv")
deg = deg.rename(columns={'comp_1': 'oBC_feature_call'})
degSig = deg[deg['p_val_adj'] < 0.1]
```

# Representative DEG Viz

```python
# subset adata
adataSub = adata[adata.obs['oBC_feature_call'].isin(['IFNA','IFNB','IFNK','IFNE','p129'])].copy()

# pull out relevant DEGs
degSub = degSig[degSig['oBC_feature_call'].isin(['GDF2','CXCL11','CCL25'])]
degSubGenes = degSub['genes'].tolist()
```

```python
sc.pl.dotplot(adataSub, degSubGenes, groupby='oBC_feature_call', dendrogram=True, layer='scaled')
```

```python
adataSub = adata[adata.obs['oBC_feature_call'].isin(['IFNA','IFNB1','IFNK','IFNE','IFNG','p129'])].copy()
sc.tl.rank_genes_groups(adataSub,'oBC_feature_call', method='wilcoxon', min_logfoldchange=0.25)
sc.tl.dendrogram(adataSub,'oBC_feature_call')
sc.pl.rank_genes_groups_dotplot(adataSub, layer='scaled', standard_scale='var', dendrogram=True, n_genes=5)
```

```python
adataSub = adata[adata.obs['oBC_feature_call'].isin(['CXCL9','CXCL10','CXCL11','p129'])].copy()
sc.tl.rank_genes_groups(adataSub,'oBC_feature_call', method='wilcoxon', min_logfoldchange=0.25)
sc.tl.dendrogram(adataSub,'oBC_feature_call')
sc.pl.rank_genes_groups_dotplot(adataSub, layer='scaled', standard_scale='var', dendrogram=True, n_genes=8)
```

```python
adataSub = adata[adata.obs['oBC_feature_call'].isin(['TGFB1','TGFB2','TGFB3','p129'])].copy()
sc.tl.rank_genes_groups(adataSub,'oBC_feature_call', method='wilcoxon', min_logfoldchange=0.25)
sc.tl.dendrogram(adataSub,'oBC_feature_call')
sc.pl.rank_genes_groups_dotplot(adataSub, layer='scaled', dendrogram=True, n_genes=8)
```
