```python
import numpy as np
import scanpy as sc
import anndata as ad
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

# Import Data

```python
adata = ad.read_h5ad("/data/rudensky/EYW/SIG04_240816/python_outs/SIG04_merge_singlets_ZscoreNorm_scanpy_clustered.h5ad")

# remove pooled samples
adata = adata[~adata.obs['oBC_feature_call'].str.contains('-pool')]
```

# Cluster Stats

```python
clusterData = adata.obs[['cluster_anno_res_1.00']]
clusterCounts = clusterData.groupby(['cluster_anno_res_1.00']).size().reset_index(name='counts')
```

```python
# Assuming your DataFrame is named `clusterCounts`
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster_anno_res_1.00', y='counts', data=clusterCounts)
plt.xticks(rotation=90)  # Rotate x labels if needed
plt.xlabel('Cluster Annotations')
plt.ylabel('Counts')
plt.title('Counts per Cluster Annotation')
plt.tight_layout()
plt.show()
```

```python
obs_data = adata.obs[['oBC_feature_call', 'leiden_res_1.00']]

# Group by oBC_feature_call and clusters, and count occurrences
grouped_data = obs_data.groupby(['oBC_feature_call', 'leiden_res_1.00']).size().reset_index(name='counts')
# Pivot the data to get clusters as columns
pivot_df = grouped_data.pivot(index='oBC_feature_call', columns='leiden_res_1.00', values='counts').fillna(0)
# Normalize the data to get frequencies (out of 1)
pivot_df = pivot_df.div(pivot_df.sum(axis=1), axis=0)

# Create the plot and set the figure size
fig, ax = plt.subplots(figsize=(30, 8))  # Adjust this to change the figure size

# Plot the stacked bar plot with normalized frequencies
pivot_df.plot(kind='bar', stacked=True, ax=ax)

# Add labels and title
ax.set_ylabel('Proportion of Clusters')
ax.set_xlabel('oBC_feature_call')
ax.set_title('Stacked Bar Plot of Cluster Proportions by oBC_feature_call')
ax.legend(title='Clusters', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
obs_data = adata.obs[['oBC_feature_call', 'leiden_res_1.00']]

# Group by oBC_feature_call and clusters, and count occurrences
grouped_data = obs_data.groupby(['oBC_feature_call', 'leiden_res_1.00']).size().reset_index(name='counts')
# Pivot the data to get clusters as columns
pivot_df = grouped_data.pivot(index='oBC_feature_call', columns='leiden_res_1.00', values='counts').fillna(0)

# Normalize the counts for each cluster by the total number of cells with that oBC_feature_call
total_cells_per_oBC = pivot_df.sum(axis=1)
normalized_counts_df = pivot_df.div(total_cells_per_oBC, axis=0)

# Calculate the log2 fold change relative to "p129"
reference_normalized_counts = normalized_counts_df.loc['p129']
log2_fold_change_df = np.log2(normalized_counts_df.div(reference_normalized_counts, axis=1)+0.1)
log2_fold_change_df = log2_fold_change_df.drop('p129')

# Melt the DataFrame to a long format for use with seaborn
melted_df = log2_fold_change_df.reset_index().melt(id_vars='oBC_feature_call', var_name='Cluster', value_name='Log2 Fold Change')

# Create the boxplot using seaborn
plt.figure(figsize=(12, 8))
ax = sns.boxplot(x='Cluster', y='Log2 Fold Change', data=melted_df, hue='Cluster', showfliers=True)

# Add labels and title
plt.ylabel('Log2(Fold Change relative to p129 + 0.1)')
plt.xlabel('Cluster')
plt.title('Log2 Fold Change of oBC_feature_call Categories by Cluster compared to p129 (with Outlier Labels)')

# Show plot
plt.tight_layout()
plt.show()
```

```python
obs_data = adata.obs[['oBC_feature_call', 'cluster_anno_res_1.00']]

# Group by oBC_feature_call and clusters, and count occurrences
grouped_data = obs_data.groupby(['oBC_feature_call', 'cluster_anno_res_1.00']).size().reset_index(name='counts')
# Pivot the data to get clusters as columns
pivot_df = grouped_data.pivot(index=['oBC_feature_call'], columns='cluster_anno_res_1.00', values='counts').fillna(0)

# Normalize the counts for each cluster by the total number of cells with that oBC_feature_call
total_cells_per_oBC = pivot_df.sum(axis=1)
normalized_counts_df = pivot_df.div(total_cells_per_oBC, axis=0)

# Calculate the log2 fold change relative to "p129"
reference_normalized_counts = normalized_counts_df.loc['p129']
# Calculate the fold change relative to "p129", replacing 0s with np.nan to avoid division by zero
fold_change_df = normalized_counts_df.div(reference_normalized_counts, axis=1).replace(0, np.nan)
log2_fold_change_df = np.log2(fold_change_df)
melted_df = log2_fold_change_df.reset_index().melt(id_vars='oBC_feature_call', var_name='Cluster', value_name='Log2 Fold Change')

# Create the boxplot using seaborn
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x='Cluster', y='Log2 Fold Change', data=melted_df, hue='Cluster', showfliers=True)
plt.ylabel('Log2(Fold Change relative to p129)')
plt.xlabel('Cluster')
plt.title('Log2 Fold Change of oBC_feature_call Categories by Cluster compared to p129 (Zero Counts Removed)')
plt.xticks(rotation=90)

# Show plot
plt.tight_layout()
plt.show()
```

```python
obs_data = adata.obs[['oBC_feature_call', 'cluster_anno_res_1.00', 'gem_group']]

# Group by oBC_feature_call, cluster_anno_res_1.00, and gem_group, and count occurrences
grouped_data = obs_data.groupby(['oBC_feature_call', 'cluster_anno_res_1.00', 'gem_group']).size().reset_index(name='counts')
pivot_df = grouped_data.pivot_table(index=['oBC_feature_call', 'gem_group'], columns='cluster_anno_res_1.00', values='counts').fillna(0)

# Split the data by lane (gem_group)
pivot_lane1 = pivot_df.xs('lane1', level='gem_group', drop_level=False)
pivot_lane2 = pivot_df.xs('lane2', level='gem_group', drop_level=False)

# Normalize oBC_feature_calls to total number of cells
total_cells_per_oBC_lane1 = pivot_lane1.sum(axis=1)
normalized_counts_lane1 = pivot_lane1.div(total_cells_per_oBC_lane1, axis=0)
total_cells_per_oBC_lane2 = pivot_lane2.sum(axis=1)
normalized_counts_lane2 = pivot_lane2.div(total_cells_per_oBC_lane2, axis=0)

# Calculate the log2 fold change relative to "p129"
reference_counts_lane1 = normalized_counts_lane1.loc['p129']
log2_fold_change_lane1 = np.log2(normalized_counts_lane1.div(reference_counts_lane1, axis=1).replace(0, np.nan))
reference_counts_lane2 = normalized_counts_lane2.loc['p129']
log2_fold_change_lane2 = np.log2(normalized_counts_lane2.div(reference_counts_lane2, axis=1).replace(0, np.nan))

# Combine the log2 fold change DataFrames for lane1 and lane2
log2_fold_change_combined = pd.concat([log2_fold_change_lane1, log2_fold_change_lane2])

# Melt the combined DataFrame for plotting
melted_df = log2_fold_change_combined.reset_index().melt(id_vars=['oBC_feature_call', 'gem_group'], var_name='Cluster', value_name='Log2 Fold Change')

# Create the boxplot using seaborn, with gem_group as the hue to compare lanes
plt.figure(figsize=(12, 8))

# Create the catplot using seaborn, with gem_group as the column for subplots
g = sns.catplot(x='Cluster', y='Log2 Fold Change', data=melted_df, kind="box",
                hue='Cluster', col='gem_group', showfliers=True, height=6, aspect=1.2)

# Set the xlabel, ylabel, and title for both subplots
g.set_axis_labels('Cluster', 'Log2(Fold Change relative to p129)')  # Set axis labels
g.set_titles('{col_name}')  # Set title based on gem_group
g.set_xticklabels(rotation=90)  # Rotate xticks for both subplot

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create a new DataFrame to store the outliers
outliers_df = pd.DataFrame(columns=melted_df.columns)

# Identify outliers based on the IQR
for cluster in melted_df['Cluster'].unique():
    cluster_data = melted_df[melted_df['Cluster'] == cluster]
    
    # For each cluster, calculate Q1 (25th percentile) and Q3 (75th percentile) and IQR
    Q1 = cluster_data['Log2 Fold Change'].quantile(0.25)
    Q3 = cluster_data['Log2 Fold Change'].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outliers as points outside of 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Get the outliers
    outliers = cluster_data[(cluster_data['Log2 Fold Change'] < lower_bound) | (cluster_data['Log2 Fold Change'] > upper_bound)]
    
    # Append the outliers to the outliers_df DataFrame
    outliers_df = pd.concat([outliers_df, outliers])
    
outliers_df.head()
```
