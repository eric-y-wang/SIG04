```python
import numpy as np
import matplotlib.pyplot as plt
import scanpy as sc
import anndata as ad
import pandas as pd
import seaborn as sns
from matplotlib.gridspec import GridSpec

sc.settings.verbosity = 3  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor="white")
```

## Import Data

```python
base_dir = "/data/rudensky/EYW/SIG04_240816/"

samples = {
    "lane1": f"{base_dir}SIG04_lane1_GEX_oBC/outs/filtered_feature_bc_matrix.h5",
    "lane2": f"{base_dir}SIG04_lane2_GEX_oBC/outs/filtered_feature_bc_matrix.h5",
}
adatasRNA = {}
adatasBC = {}

for sample_id, filename in samples.items():
    sample_adata = sc.read_10x_h5(filename, gex_only=False) 
    sample_adata.var_names_make_unique()
    adatasRNA[sample_id] = sample_adata[:, sample_adata.var["feature_types"] == "Gene Expression"].copy()
    adatasBC[sample_id] = sample_adata[:, sample_adata.var["feature_types"] == "CRISPR Guide Capture"].copy()
```

## QC Metrics

Here we will look at QC values for the transcriptome and barcodes. I will add the QC info on barcodes directly to the RNA anndata objects.

```python
for sample_id in adatasRNA:
    # mitochondrial genes
    adatasRNA[sample_id].var["mt"] = adatasRNA[sample_id].var_names.str.startswith("mt-")
    # ribosomal genes
    adatasRNA[sample_id].var["ribo"] = adatasRNA[sample_id].var_names.str.startswith(("Rps", "Rpl"))
    # calculate qc metrics
    sc.pp.calculate_qc_metrics(
        adatasRNA[sample_id], qc_vars=["mt", "ribo"], inplace=True, log1p=True
    )
    # Calculate the total counts for each cell for sparse matrix
    adatasRNA[sample_id].obs['oBC_total_counts'] = adatasBC[sample_id].X.sum(axis=1).A1
    # Calculate the log1p of the total counts
    adatasRNA[sample_id].obs['oBC_log1p_total_counts'] = np.log1p(adatasBC[sample_id].obs['oBC_total_counts'])
```

```python
# print QC metrics for each sample
for sample_id in adatasRNA:
    print(sample_id)
    # Plot the violin plots for the subsetted data
    sc.pl.violin(
        adatasRNA[sample_id],
        ["n_genes_by_counts", "total_counts", "pct_counts_mt"],
        jitter=0.4,
        multi_panel=True
    )
```

```python
# Assume adatasRNA is your list of AnnData objects
qcValues = [1500, 2000, 5]  # [min_genes, min_counts, max_percent_mt]

# Initialize a list to store metadata from all samples
all_metadata = []

# Iterate over the list of AnnData objects
for sample_id in adatasRNA:
    # Extract relevant metadata
    metadata = adatasRNA[sample_id].obs.copy()
    metadata['sample'] = sample_id
    metadata['log10GenesPerUMI'] = metadata['n_genes_by_counts'].apply(lambda x: np.log10(x))/ metadata['total_counts'].apply(lambda x: np.log10(x))
    
    # Append to the list of metadata
    all_metadata.append(metadata)

# Concatenate all metadata into a single DataFrame
combined_metadata = pd.concat(all_metadata)

# Set up the figure layout
fig = plt.figure(figsize=(25, 20))
gs = GridSpec(3, 3, figure=fig)

# Plot 1: UMI count density
ax1 = fig.add_subplot(gs[0, 0])
sns.kdeplot(data=combined_metadata, x='total_counts', hue='sample', fill=True, alpha=0.2, ax=ax1)
ax1.set_xscale('log')
ax1.set_ylabel("Cell density")
ax1.axvline(x=qcValues[1], color='red')
ax1.set_title("UMI Counts per Cell")

# Plot 2: Genes detected density
ax2 = fig.add_subplot(gs[0, 1])
sns.kdeplot(data=combined_metadata, x='n_genes_by_counts', hue='sample', fill=True, alpha=0.2, ax=ax2)
ax2.set_xscale('log')
ax2.axvline(x=qcValues[0], color='red')
ax2.set_title("Genes Detected per Cell")

# Plot 3: Mitochondrial gene expression density
ax3 = fig.add_subplot(gs[0, 2])
sns.kdeplot(data=combined_metadata, x='pct_counts_mt', hue='sample', fill=True, alpha=0.2, ax=ax3)
ax3.axvline(x=qcValues[2], color='red')
ax3.set_xlim(0, 10)
ax3.set_title("Mitochondrial Gene Expression per Cell")

# Plot 4: Genes detected per UMI density
ax4 = fig.add_subplot(gs[1, 0])
sns.kdeplot(data=combined_metadata, x='log10GenesPerUMI', hue='sample', fill=True, alpha=0.2, ax=ax4)
ax4.axvline(x=0.8, color='red')
ax4.set_title("Genes Detected per UMI")

# Plot 5: Correlation between genes detected and UMI counts
ax5 = fig.add_subplot(gs[1, 1])
sns.scatterplot(data=combined_metadata, x='total_counts', y='n_genes_by_counts', hue='pct_counts_mt', palette="viridis", ax=ax5)
ax5.set_xscale('log')
ax5.set_yscale('log')
ax5.axvline(x=qcValues[1], color='red')
ax5.axhline(y=qcValues[0], color='red')
ax5.set_title("Genes Detected vs UMI Counts")

# Plot 6: Filtered correlation plot
ax6 = fig.add_subplot(gs[1, 2])
filtered_metadata = combined_metadata[combined_metadata['pct_counts_mt'] < qcValues[2]]
sns.scatterplot(data=filtered_metadata, x='total_counts', y='n_genes_by_counts', hue='pct_counts_mt', palette="viridis", ax=ax6)
ax6.set_xscale('log')
ax6.set_yscale('log')
ax6.axvline(x=qcValues[1], color='red')
ax6.axhline(y=qcValues[0], color='red')
ax6.set_title("Filtered Genes Detected vs UMI Counts")

# Plot 7: oBC UMI count density
ax7 = fig.add_subplot(gs[2, 0])
sns.kdeplot(data=combined_metadata, x='oBC_log1p_total_counts', hue='sample', fill=True, alpha=0.2, ax=ax7)
ax7.set_ylabel("Cell density")
ax7.set_title("oBC UMI Counts per Cell")

# Plot 5: Correlation between oBC and umi counts
ax8 = fig.add_subplot(gs[2, 1])
sns.scatterplot(data=combined_metadata, x='oBC_total_counts', y='total_counts', hue='pct_counts_mt', palette="viridis", ax=ax8)
ax8.set_title("oBC vs RNA UMI Counts")
ax8.set_xscale('log')
ax8.set_yscale('log')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
